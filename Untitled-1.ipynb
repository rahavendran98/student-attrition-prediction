Feature Engineering

-- Creating new features which will be helpful to built a model


```python

df['academic_risk'] = (
    (10 - df['current_gpa']) +
    (100 - df['attendance_percentage']) / 10 +
    df['backlogs']
)
print(df['academic_risk'].describe())
```

```python

df['financial_risk'] = (
    df['financial_stress_score'] +
    (df['fee_payment_delay'] == 'Yes').astype(int) * 5 +
    (df['part_time_job'] == 'Yes').astype(int) * 2
)
print(df['financial_risk'].describe())
```

```python
df['engagement_score'] = (
    df['class_participation_score'] +
    df['student_satisfaction_score'] +
    df['peer_interaction_score']
) / 3
print(df['engagement_score'].describe())
```

```python
df['support_score'] = df['counseling_sessions_attended'] + df['mentor_meetings']
print(df['support_score'].describe())
```

```python

def age_group(age):
    if age <= 19:
        return 'Young'     
    elif age <= 22:
        return 'Mid'         
    else:
        return 'Older'       

df['age_group'] = df['age'].apply(age_group)
print(df['age_group'].value_counts())

```

```python
# Check correlation of new features with target
new_features = ['academic_risk', 'financial_risk', 'engagement_score', 'support_score']
print("Correlation of new features with dropout:\n")
print(df[new_features + ['dropout']].corr()['dropout'].sort_values(ascending=False))
```

## Feature Engineering Insights 


-- Four new features were created to better understand student behavior and risk patterns.

-- academic_risk became the strongest predictor of dropout in the dataset.

-- Students with high academic risk showed much higher dropout rates compared to low-risk students.

-- The combined academic_risk feature performed better than individual academic features like backlogs alone.

-- Students with lower engagement levels had slightly higher dropout chances.

-- support_score showed very weak impact, possibly because students usually seek support only after facing difficulties.

-- Age groups showed almost no major difference in dropout rates.

-- Students who dropped out had noticeably higher academic risk compared to students who continued.

-- Overall, academic_risk is the most important feature, while other engineered features may still help advanced models find hidden patterns.

```python
binary_yes_no = [
    'first_generation_learner', 'extracurricular_participation', 'scholarship_status',
    'fee_payment_delay', 'part_time_job', 'hostel_status', 'health_issues_reported'
]

df[binary_yes_no] = (df[binary_yes_no] == 'Yes').astype(int)

print("Binary columns encoded")
print(df[binary_yes_no].head())
```

-- binary categorical coloumns have been encoded into numerical values

```python
ordinal_mappings = {
    'family_income_level': {'Low': 0, 'Middle': 1, 'High': 2},
    'parent_education': {'No Formal': 0, 'School': 1, 'Graduate': 2, 'Postgraduate': 3},
    'age_group': {'Young': 0, 'Mid': 1, 'Older': 2}
}

for col, mapping in ordinal_mappings.items():
    df[col] = df[col].map(mapping)

print("Ordinal columns encoded")
print(df[list(ordinal_mappings)].head())
```

Nominal columns to one hot encoding

```python
nominal_cols = ['gender', 'location', 'course_type', 'admission_type']
df = pd.get_dummies(df, columns=nominal_cols, drop_first=True)

print(f"One-hot encoding done. New shape: {df.shape}")
print("New columns created:")
print([c for c in df.columns if c.startswith(tuple(f"{nc}_" for nc in nominal_cols))])
```
